#### 1) . What is the difference between enclosing a list comprehension in square brackets and parentheses?
**Ans:** both square brackets and parentheses can be used to enclose a list comprehension, but they have different meanings and produce different types of objects.

When a list comprehension is enclosed in square brackets, it creates a new list object that contains the results of the comprehension

When a list comprehension is enclosed in parentheses, it creates a new generator object that generates the results of the comprehension on the fly as needed

The key difference between using square brackets and parentheses to enclose a list comprehension is that square brackets create a new list object, while parentheses create a new generator object. Depending on the use case, one or the other may be more appropriate. If you need to access the results of the comprehension multiple times or modify them in place, a list comprehension may be more appropriate. If you only need to iterate over the results once or want to conserve memory, a generator expression enclosed in parentheses may be more appropriate.

```python
squares = [x*x for x in range(1, 6)]
print(squares)

squares = (x*x for x in range(1, 6))
print(squares)

```

#### 2) What is the relationship between generators and iterators?
**Ans:** Generators and iterators are related concepts in Python, as generators are a type of iterator. Both generators and iterators are used to iterate over a sequence of values, but they differ in how they generate those values.

Iterators are objects that implement the iterator protocol, which consists of the __iter__() and __next__() methods. The __iter__() method returns the iterator object itself, and the __next__() method returns the next value in the sequence or raises the StopIteration exception when there are no more values.

Generators, on the other hand, are a type of iterator that are defined using a function that contains the yield keyword. When a generator function is called, it returns a generator object that can be used to iterate over the values generated by the function. Each time the yield keyword is encountered in the function, the current value is returned to the caller and the state of the function is saved. The next time the generator is iterated, the function resumes execution from where it left off and generates the next value.

```python
# Example of iterartor
iter_str = iter(['iNeuron','Full','Stack','Data Science'])
print(type(iter_str))
print(next(iter_str))
print(next(iter_str))
print(next(iter_str))
print(next(iter_str))
print(iter_str)  

# Example of Generator
def cube_numbers(in_num):
    for ele in range(in_num+1):
        yield ele**3

out_num = cube_numbers(4)
print(next(out_num))
print(next(out_num))
print(next(out_num))
print(next(out_num))
print(next(out_num))
```

#### 3) What are the signs that a function is a generator function?
**Ans:** A function is a generator function if it contains the yield keyword. The yield keyword is used to generate a sequence of values on the fly, as opposed to returning a single value like a regular function.

Here are some signs that a function is a generator function:

The function contains the yield keyword
The function returns a generator object instead of a regular value
The function is usually called with a for loop or the next() function to iterate over the sequence of values generated by the function
The function may use a loop or other control structures to generate the sequence of values, but it must contain at least one yield statement

```python
def generate_even_numbers(n):
    i = 0
    while i < n:
        yield i * 2
        i += 1
```

#### 4) What is the purpose of a yield statement?
**Ans:** The yield statement is used in Python to create a generator function that generates a sequence of values on the fly. It allows a function to return a value and then continue from where it left off when called again, instead of starting over from the beginning like a regular function.

When a yield statement is encountered in a generator function, it returns the specified value to the caller and saves the state of the function so that it can resume execution from where it left off the next time it is called. Each time the function is called, it generates the next value in the sequence until there are no more values to generate.
